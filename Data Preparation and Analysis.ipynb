-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.

3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here

# Splitting the 'genre' column by the pipe '|'
df_split_genre = movies_after_colrowdrop_dropdup['genres'].str.split('|', expand=True)
print("Dataframe extracted from 'genre' by splitting with pipe '|' :")
display(df_split_genre)

print("\n---------------------------------------------------------------------------------------------------\n")

# Extracting and renaming the first two columns of dataframe
df_split_genre = df_split_genre.iloc[:,0:2]
df_split_genre.rename(columns = {0:'genre_1', 1:'genre_2'}, inplace = True) 
print("Dataframe after extracting first two columns :")
display(df_split_genre)

print("\n---------------------------------------------------------------------------------------------------\n")

# For movies with single genre, replacing the null values in 'genre_2' with the value from 'genre_1'
df_split_genre.genre_2.fillna(df_split_genre.genre_1, inplace=True)
print("Dataframe after replacing null values in 'genre_2' with value from 'genre_1'")
display(df_split_genre)

print("\n---------------------------------------------------------------------------------------------------\n")

# Appending the newly created genre columns to the existing data frame
movies_after_colrowdrop_dropdup_split = movies_after_colrowdrop_dropdup.copy(deep=True)
movies_after_colrowdrop_dropdup_split[['genre_1', 'genre_2']] = df_split_genre[['genre_1', 'genre_2']]
print("Existing dataframe after appending the newly created columns :")
display(movies_after_colrowdrop_dropdup_split[['movie_title', 'director_name', 'genres', 'genre_1', 'genre_2']])
```

```python
# Grouping the dataframe using 'genre_1' as primary colummn and 'genre_2' as secondary column

movies_by_segment = movies_after_colrowdrop_dropdup_split.groupby(['genre_1','genre_2'])# Write your code for grouping the dataframe here
print("Grouping dataframe using 'genre_1' as primary colummn and 'genre_2' as secondary column :")
display(movies_by_segment.describe())
```

```python
# To find out the 5 most popular combo by finding the mean of 'gross' values

PopGenre = movies_by_segment.mean().nlargest(5, columns='gross')# Write your code for getting the 5 most popular combo of genres here
print("5 most popular combo of genre by considering the mean of 'gross' values :")
display(PopGenre[['gross']])
```

```python
# # Write the code for calculating decade here

# Function for calculating decade based on year
def set_decade(year):
    if year is not None:
        temp1 = year // 10
        temp1_str = str(temp1)
        return temp1_str + '0s'
    else:
        return None
    
# Converting type of 'title_year' from float64 to int
movies_after_colrowdrop_dropdup['title_year'] = movies_after_colrowdrop_dropdup['title_year'].astype('int')

# Apply function for each record and store it in 'Decade' column
movies_after_colrowdrop_dropdup['decade'] = movies_after_colrowdrop_dropdup.apply(lambda x: set_decade(x['title_year']), axis = 1)
print('Dataframe after calculating decade column based on year :')
display(movies_after_colrowdrop_dropdup[['movie_title', 'director_name', 'title_year', 'decade']])
```
